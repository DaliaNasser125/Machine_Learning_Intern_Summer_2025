```python
data.dropna(axis=0,inplace=True)
```

```python
def remove_outliers_iqr(data, column):

    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
```

```python
for col in data.select_dtypes(include='number').columns:
    data = remove_outliers_iqr(data, col)
```

```python
import matplotlib.pyplot as plt

# List of airport distance columns
airport_columns = ['jfk_dist', 'ewr_dist', 'lga_dist', 'sol_dist', 'nyc_dist']

# Bin width
bin_width = 0.125

# Create subplots
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(15, 12))
axs = axs.flatten()

# Loop through each airport distance column
for i, col in enumerate(airport_columns):
    # Bin the distance
    data[f'{col}_bin'] = (data[col] // bin_width) * bin_width

    # Group by binned distance and calculate average fare
    avg_fares = data.groupby(f'{col}_bin')['fare_amount'].mean().reset_index()

    # Round bin values for prettier x-axis
    avg_fares[f'{col}_bin'] = avg_fares[f'{col}_bin'].round(2)

    # Plot
    axs[i].plot(avg_fares[f'{col}_bin'], avg_fares['fare_amount'], marker='o', linestyle='-', color='green')
    axs[i].set_title(f'Avg Fare vs Distance to {col.upper().replace("_DIST", "")}')
    axs[i].set_xlabel('Distance (km)')
    axs[i].set_ylabel('Average Fare ($)')
    axs[i].grid(True)

# Remove any unused subplot if there are fewer than 6
if len(airport_columns) < len(axs):
    for j in range(len(airport_columns), len(axs)):
        fig.delaxes(axs[j])

plt.tight_layout()
plt.show()

```

```python


# Bin the distances (e.g., every 0.25 km)
bin_width = 0.125
data['distance_bin'] = (data['distance'] // bin_width) * bin_width

# Group by bin and compute average fare
avg_fares = data.groupby('distance_bin')['fare_amount'].mean().reset_index()

# Plot
plt.figure(figsize=(10, 6))
plt.plot(avg_fares['distance_bin'], avg_fares['fare_amount'], marker='o', linestyle='-', color='green')
plt.xlabel('Distance')
plt.ylabel('Average Fare Amount ($)')
plt.title('Average Fare Amount vs Distance')
plt.grid(True)
plt.show()
```
