```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/final_internship_data.csv')
data.head()
```

```python
print(data.info())
```

```python
print(data.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.histplot(data['fare_amount'], bins=50, kde=True)
plt.title('Distribution of Fare Amount')
plt.xlabel('Fare Amount')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='passenger_count', data=data)
plt.title('Passenger Count Distribution')
plt.xlabel('Passenger Count')
plt.ylabel('Number of Rides')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='passenger_count', y='fare_amount', data=data)
plt.title('Fare Amount vs Passenger Count')
plt.xlabel('Passenger Count')
plt.ylabel('Fare Amount')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='distance', y='fare_amount', data=data, alpha=0.5)
plt.title('Fare Amount vs Trip Distance')
plt.xlabel('Distance (km)')
plt.ylabel('Fare Amount')
plt.grid(True)
plt.show()
```

```python
for column in data.columns:
    unique_vals = data[column].unique()
    print(f"عمود '{column}' يحتوي على {len(unique_vals)} قيمة فريدة:")
    print(unique_vals)
    print("-" * 50)
```

```python
corr = data.corr(numeric_only = True)
plt.figure(figsize = (14, 12))
sns.heatmap(corr, cmap = 'coolwarm', annot = True, linewidth = 0.5)
plt.show()
```
