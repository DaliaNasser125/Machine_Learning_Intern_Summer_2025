```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\mahmoud\Desktop\Cellula intern\final_internship_data.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
#Q1. How does the fare amount vary across different car conditions under various weather types?
plt.figure(figsize=(10, 6))
sns.boxplot(x='Car Condition', y='fare_amount', hue='Weather', data=df)
plt.title("Fare Amount by Car Condition and Weather")
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#Q2. What is the hourly distribution of rides under different traffic conditions?
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='hour', hue='Traffic Condition')
plt.title("Number of Rides per Hour by Traffic Condition")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Rides")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#Q3. What are the average fare trends across the week and different hours of the day?
pivot_fare = df.pivot_table(index='weekday', columns='hour', values='fare_amount', aggfunc='mean')
plt.figure(figsize=(14, 6))
sns.heatmap(pivot_fare, cmap='YlGnBu', annot=False)
plt.title("Heatmap of Average Fare by Weekday and Hour")
plt.xlabel("Hour")
plt.ylabel("Weekday (0=Monday, 6=Sunday)")
plt.tight_layout()
plt.show()
```

```python
#Q4.How do average fares change throughout the year under different weather conditions?
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='month', y='fare_amount', hue='Weather', ci=None)
plt.title("Monthly Trend of Average Fare by Weather")
plt.xlabel("Month")
plt.ylabel("Average Fare")
plt.tight_layout()
plt.show()
```

```python
#Q5.At which hours and weekdays are rides most frequent?
ride_counts = df.pivot_table(index='weekday', columns='hour', values='User ID', aggfunc='count')
plt.figure(figsize=(14, 6))
sns.heatmap(ride_counts, cmap='magma')
plt.title("Ride Count by Weekday and Hour")
plt.xlabel("Hour")
plt.ylabel("Weekday")
plt.tight_layout()
plt.show()
```

```python
#Q6.How do average monthly fares change with traffic conditions?
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='month', y='fare_amount', hue='Traffic Condition', ci=None)
plt.title("Monthly Average Fare by Traffic Condition")
plt.tight_layout()
plt.show()
```

```python
#Q7.How does passenger count vary by weather ?
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='passenger_count', hue='Weather')
plt.title("Passenger Count by Weather")
plt.tight_layout()
plt.show()
```

```python
#Q8.Does weather affect average travel distance over months?
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='month', y='distance', hue='Weather', ci=None)
plt.title("Monthly Average Distance by Weather")
plt.tight_layout()
plt.show()
```

```python
#Q9.How does total trip distance (distance) relate to distance from NYC center?
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='nyc_dist', y='distance', alpha=0.4)
plt.title("Trip Distance vs NYC Distance")
plt.xlabel("NYC Distance")
plt.ylabel("Trip Distance")
plt.tight_layout()
plt.show()
```

```python
#Q10.How do airport distances differ across weather conditions?
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Weather', y='jfk_dist')
plt.title("JFK Distance by Weather Condition")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Q11.Which landmark distance has the strongest correlation with fare?
dist_cols = ['jfk_dist', 'ewr_dist', 'lga_dist', 'sol_dist', 'nyc_dist', 'distance', 'fare_amount']
corr = df[dist_cols].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Distance Columns and Fare")
plt.tight_layout()
plt.show()
```

```python
#Q12. Do trips closer to NYC tend to cost less â€” regardless of distance?
df['distance_bin'] = pd.cut(df['distance'], bins=[0, 2, 5, 10, 20, 50], labels=["0â€“2km", "2â€“5km", "5â€“10km", "10â€“20km", "20km+"])

g = sns.FacetGrid(df, col='distance_bin', col_wrap=3, height=4)
g.map(sns.scatterplot, 'nyc_dist', 'fare_amount', alpha=0.3)
g.fig.suptitle("Fare vs NYC Distance by Trip Distance Group", y=1.05)
plt.tight_layout()
plt.show()
```

```python
#Q13.How does the average fare evolve over the years?
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='year', y='fare_amount', estimator='mean', ci=None, marker='o')
plt.title("Yearly Trend of Average Fare")
plt.xlabel("Year")
plt.ylabel("Average Fare Amount")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Q13.What is the distribution of traffic conditions in the dataset?
traffic_counts = df['Traffic Condition'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(traffic_counts, labels=traffic_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))
plt.title("Distribution of Traffic Conditions")
plt.tight_layout()
plt.show()
```

```python
#Q14.What is the proportion of rides taken in each weather condition?
weather_counts = df['Weather'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(weather_counts, labels=weather_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title("Distribution of Rides by Weather Condition")
plt.tight_layout()
plt.show()
```

```python
#Q15.How does the number of passengers affect the average fare during different hours of the day?
grouped = df.groupby(['hour', 'passenger_count'])['fare_amount'].mean().reset_index()

plt.figure(figsize=(14, 6))
sns.barplot(data=grouped, x='hour', y='fare_amount', hue='passenger_count', palette='Spectral')
plt.title("Average Fare by Hour and Passenger Count")
plt.xlabel("Hour of Day")
plt.ylabel("Average Fare Amount")
plt.legend(title="Passenger Count", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
