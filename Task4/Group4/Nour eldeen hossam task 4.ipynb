```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\\Users\\noorh\\Downloads\\final_internship_data.csv")
```

```python
weekday_map = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
df['weekday_name'] = df['weekday'].map(weekday_map)
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python

plt.figure()
sns.histplot(df['fare_amount'], bins=100, kde=False, color='blue')
plt.title('Distribution of Fare Amount', fontsize=16)
plt.xlabel('Fare ($)', fontsize=12)
plt.ylabel('Number of Trips', fontsize=12)
plt.xlim(0, 100)
plt.show()
```

```python

plt.figure()
sns.scatterplot(x='distance', y='fare_amount', data=df.sample(50000, random_state=42), alpha=0.5, s=10)
plt.title('Distance vs. Fare Amount', fontsize=16)
plt.xlabel('Distance (km)', fontsize=12)
plt.ylabel('Fare ($)', fontsize=12)
plt.xlim(0, 100)
plt.ylim(0, 300)
plt.show()
```

```python
plt.figure()
trips_by_hour = df.groupby('hour').size()
sns.lineplot(x=trips_by_hour.index, y=trips_by_hour.values, marker='o', color='purple')
plt.title('Total Number of Trips by Hour of the Day', fontsize=16)
plt.xlabel('Hour of the Day (0-23)', fontsize=12)
plt.ylabel('Total Number of Trips', fontsize=12)
plt.xticks(range(0, 24))
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
plt.figure()
sns.countplot(x='weekday_name', data=df, order=weekday_order, palette='viridis')
plt.title('Total Trips per Day of the Week', fontsize=16)
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Total Number of Trips', fontsize=12)
plt.show()

```

```python
plt.figure()
sns.countplot(x='passenger_count', data=df, palette='plasma')
plt.title('Distribution of Passenger Count per Trip', fontsize=16)
plt.xlabel('Number of Passengers', fontsize=12)
plt.ylabel('Number of Trips', fontsize=12)
plt.savefig("5_passenger_count_distribution.png")
plt.close()
```

```python
plt.figure()
sns.boxplot(x='Weather', y='fare_amount', data=df, palette='coolwarm')
plt.title('Fare Amount Distribution by Weather Condition', fontsize=16)
plt.xlabel('Weather Condition', fontsize=12)
plt.ylabel('Fare ($)', fontsize=12)
plt.ylim(0, 75)
plt.show()

```

```python
plt.figure()
sns.boxplot(x='Traffic Condition', y='fare_amount', data=df, palette='magma')
plt.title('Fare Amount Distribution by Traffic Condition', fontsize=16)
plt.xlabel('Traffic Condition', fontsize=12)
plt.ylabel('Fare ($)', fontsize=12)
plt.ylim(0, 75)
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Car Condition', y='fare_amount', data=df, palette='Set2')
plt.title('Fare Amount Distribution by Car Condition', fontsize=16)
plt.xlabel('Car Condition', fontsize=12)
plt.ylabel('Fare ($)', fontsize=12)
plt.ylim(0, 75)
plt.show()
```
