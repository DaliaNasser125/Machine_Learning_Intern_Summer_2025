```python
import pandas as pd

df = pd.read_csv('final_internship_data.csv')
```

```python
display(df.head())
df.info()
display(df.describe())
```

```python
import matplotlib.pyplot as plt

numerical_cols = ['fare_amount', 'passenger_count', 'distance', 'hour', 'day', 'month', 'year']

for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    df[col].hist(bins=50)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
categorical_cols = ['Car Condition', 'Weather', 'Traffic Condition']

for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Frequency of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['distance'], df['fare_amount'], alpha=0.5)
plt.title('Fare Amount vs. Distance')
plt.xlabel('Distance')
plt.ylabel('Fare Amount')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df.boxplot(column='fare_amount', by='passenger_count')
plt.title('Fare Amount by Passenger Count')
plt.xlabel('Passenger Count')
plt.ylabel('Fare Amount')
plt.suptitle('') # Suppress the default suptitle
plt.show()
```
