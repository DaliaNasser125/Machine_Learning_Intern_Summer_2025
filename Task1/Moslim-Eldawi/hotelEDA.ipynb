```python
# Load the dataset
df = pd.read_csv('hotel_booking.csv')
```

```python
## 1. Initial Data Exploration
print("Dataset Shape:", df.shape)
print("\nFirst 5 rows:")
display(df.head())
print("\nData Types:")
print(df.dtypes)
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Basic statistics
print("\nDescriptive Statistics:")
display(df.describe(include='all'))
```

```python
## 2. Data Cleaning and Preparation - Improved Version

# First attempt date conversion with error reporting
try:
    df['date of reservation'] = pd.to_datetime(df['date of reservation'], format='%m/%d/%Y')
except ValueError as e:
    print(f"Encountered error: {e}")
    print("\nAttempting to identify problematic rows...")
```

```python

# Proceed with the rest of the date-based features
df['reservation_year'] = df['date of reservation'].dt.year
df['reservation_month'] = df['date of reservation'].dt.month
df['reservation_day'] = df['date of reservation'].dt.day
```

```python

# Create total nights column
df['total_nights'] = df['number of weekend nights'] + df['number of week nights']
```

```python

# Create total guests column
df['total_guests'] = df['number of adults'] + df['number of children']
```

```python

# Create cancellation rate column
df['cancellation_rate'] = df['P-C'] / (df['P-C'] + df['P-not-C'])
df['cancellation_rate'] = df['cancellation_rate'].fillna(0)
```

```python
## 3. Univariate Analysis
# Numerical Features
num_cols = ['number of adults', 'number of children', 'number of weekend nights',
            'number of week nights', 'lead time', 'average price', 'special requests',
            'total_nights', 'total_guests']

plt.figure(figsize=(15, 10))
for i, col in enumerate(num_cols):
    plt.subplot(3, 3, i+1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python

# Categorical Features
cat_cols = ['type of meal', 'car parking space', 'room type', 'market segment type',
            'booking status', 'reservation_year', 'reservation_month']

plt.figure(figsize=(15, 15))
for i, col in enumerate(cat_cols):
    plt.subplot(3, 3, i+1)
    if df[col].nunique() > 10:
        sns.countplot(y=df[col], order=df[col].value_counts().index[:10])
    else:
        sns.countplot(x=df[col])
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

## 4. Bivariate Analysis
# Cancellation rate by different features
plt.figure(figsize=(15, 8))
sns.barplot(x='room type', y='cancellation_rate', data=df, ci=None)
plt.title('Cancellation Rate by Room Type')
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.boxplot(x='booking status', y='lead time', data=df)
plt.title('Lead Time Distribution by Booking Status')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.boxplot(x='booking status', y='average price', data=df)
plt.title('Average Price Distribution by Booking Status')
plt.show()
```

```python

# Correlation matrix
plt.figure(figsize=(12, 8))
corr = df[['number of adults', 'number of children', 'number of weekend nights',
           'number of week nights', 'lead time', 'average price', 'special requests',
           'total_nights', 'total_guests']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```

```python

## 5. Time Series Analysis
# Bookings over time
monthly_bookings = df.groupby(['reservation_year', 'reservation_month'])['Booking_ID'].count().reset_index()
monthly_bookings['date'] = pd.to_datetime(monthly_bookings['reservation_year'].astype(str) + '-' +
                           monthly_bookings['reservation_month'].astype(str) + '-01')

plt.figure(figsize=(15, 6))
sns.lineplot(x='date', y='Booking_ID', data=monthly_bookings)
plt.title('Monthly Bookings Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.show()
```

```python

# Cancellations over time
monthly_cancellations = df[df['booking status'] == 'Canceled'].groupby(
    ['reservation_year', 'reservation_month'])['Booking_ID'].count().reset_index()
monthly_cancellations['date'] = pd.to_datetime(monthly_cancellations['reservation_year'].astype(str) + '-' +
                                monthly_cancellations['reservation_month'].astype(str) + '-01')

plt.figure(figsize=(15, 6))
sns.lineplot(x='date', y='Booking_ID', data=monthly_cancellations)
plt.title('Monthly Cancellations Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cancellations')
plt.xticks(rotation=45)
plt.show()
```

```python

## 6. Customer Behavior Analysis
# Repeat customers vs. new customers
repeat_customers = df[df['repeated'] > 0]
new_customers = df[df['repeated'] == 0]

print(f"Repeat customers: {len(repeat_customers)} ({len(repeat_customers)/len(df)*100:.2f}%)")
print(f"New customers: {len(new_customers)} ({len(new_customers)/len(df)*100:.2f}%)")
```

```python

plt.figure(figsize=(10, 6))
sns.countplot(x='repeated', data=df)
plt.title('Distribution of Customer Repeat Visits')
plt.xlabel('Number of Repeat Visits')
plt.ylabel('Count')
plt.show()
```

```python

# Cancellation rate by customer type
repeat_cancel_rate = repeat_customers['booking status'].value_counts(normalize=True)['Canceled']
new_cancel_rate = new_customers['booking status'].value_counts(normalize=True)['Canceled']

plt.figure(figsize=(8, 6))
sns.barplot(x=['Repeat Customers', 'New Customers'], y=[repeat_cancel_rate, new_cancel_rate])
plt.title('Cancellation Rate by Customer Type')
plt.ylabel('Cancellation Rate')
plt.show()
```

```python

## 7. Advanced Analysis
# RFM Analysis (Recency, Frequency, Monetary)
rfm = df.groupby('Booking_ID').agg({
    'date of reservation': lambda x: (df['date of reservation'].max() - x.max()).days,
    'repeated': 'max',
    'average price': 'mean'
}).rename(columns={
    'date of reservation': 'recency',
    'repeated': 'frequency',
    'average price': 'monetary'
})
```

```python

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.histplot(rfm['recency'], bins=50)
plt.title('Recency Distribution')
```

```python

plt.subplot(1, 3, 2)
sns.histplot(rfm['frequency'], bins=50)
plt.title('Frequency Distribution')
```

```python

plt.subplot(1, 3, 3)
sns.histplot(rfm['monetary'], bins=50)
plt.title('Monetary Distribution')

plt.tight_layout()
plt.show()
```

```python

# Customer segmentation by booking behavior
df['customer_segment'] = pd.cut(df['repeated'],
                               bins=[-1, 0, 1, 3, float('inf')],
                               labels=['New', 'Occasional', 'Regular', 'Frequent'])

plt.figure(figsize=(10, 6))
sns.countplot(x='customer_segment', hue='booking status', data=df)
plt.title('Booking Status by Customer Segment')
plt.show()
```

```python

## 8. Key Insights Summary
print("\n=== KEY INSIGHTS ===")
# Overall cancellation rate
cancel_rate = df['booking status'].value_counts(normalize=True)['Canceled']
print(f"1. Overall Cancellation Rate: {cancel_rate*100:.2f}%")
```

```python

# Most popular room type
popular_room = df['room type'].value_counts().index[0]
print(f"2. Most Popular Room Type: {popular_room}")
```

```python

# Average lead time difference between canceled and not canceled
avg_lead_diff = df.groupby('booking status')['lead time'].mean().diff().iloc[-1]
print(f"3. Canceled bookings have {avg_lead_diff:.2f} more days lead time on average")
```

```python

# Average price difference between canceled and not canceled
avg_price_diff = df.groupby('booking status')['average price'].mean().diff().iloc[-1]
print(f"4. Canceled bookings are ${abs(avg_price_diff):.2f} {'higher' if avg_price_diff > 0 else 'lower'} on average")
```

```python

# Most common booking channel
common_channel = df['market segment type'].value_counts().index[0]
print(f"5. Most Common Booking Channel: {common_channel}")
```

```python

# Month with highest cancellations
monthly_cancel = df[df['booking status'] == 'Canceled'].groupby('reservation_month')['Booking_ID'].count()
highest_cancel_month = monthly_cancel.idxmax()
print(f"6. Month with Highest Cancellations: {highest_cancel_month} ({monthly_cancel.max()} cancellations)")
```
