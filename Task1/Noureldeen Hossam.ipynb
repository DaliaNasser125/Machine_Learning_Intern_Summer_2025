```python
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r"C:\Users\noorh\Downloads\first inten project.csv" 

try:
    df = pd.read_csv(file_path)

    df['date of reservation'] = pd.to_datetime(df['date of reservation'], format='mixed', errors='coerce')

    df['reservation_year'] = df['date of reservation'].dt.year
    df['reservation_month'] = df['date of reservation'].dt.month_name()
    df['reservation_day_of_week'] = df['date of reservation'].dt.day_name()
    print(df[['date of reservation', 'reservation_year', 'reservation_month', 'reservation_day_of_week']].head())
    df.info()

except FileNotFoundError:
    print('not found')
```

```python
booking_status_counts = df['booking status'].value_counts()

booking_status_percentages = df['booking status'].value_counts(normalize=True) * 100

print(booking_status_counts)
print(booking_status_percentages.round(2))
plt.figure(figsize=(8, 8))
plt.pie(booking_status_percentages, labels=booking_status_percentages.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title(' Booking Status Distribution', fontsize=16)
plt.show()
```

```python
repeated_customers_counts = df['repeated'].value_counts()
repeated_customers_percentages = df['repeated'].value_counts(normalize=True) * 100

print("Slide 2: Customer Behavior: Are They Repeat Visitors?")
print("Number of bookings by customer type (repeat vs. non-repeat):")
print(repeated_customers_counts)
print("\nPercentage of bookings by customer type:")
print(repeated_customers_percentages.round(2))
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='repeated', hue='booking status', palette='viridis')
plt.title(' Booking Status of repeated and non repeated customers', fontsize=16)
plt.xlabel('Is the Customer a Repeat Visitor? ( No: New Customer , Yes: Repeat Customer )', fontsize=12)
plt.ylabel('Number of Bookings', fontsize=12)
plt.xticks(ticks=[0, 1], labels=['New Customer', 'Repeat Customer'])
plt.show()
```

```python

print('Slide 3: Repeat Customer Behavior (P-C=1 vs P-not-C=1)')

total_repeat = (df['repeated'] == 1).sum()

pc_1_count = ((df['repeated'] == 1) & (df['P-C'] == 1)).sum()
pnotc_1_count = ((df['repeated'] == 1) & (df['P-not-C'] == 1)).sum()

pc_1_percent = (pc_1_count / total_repeat) * 100 if total_repeat > 0 else 0
pnotc_1_percent = (pnotc_1_count / total_repeat) * 100 if total_repeat > 0 else 0

print(f"Percentage of Repeat Customers with P-C=1: {pc_1_percent:.2f}%")
print(f"Percentage of Repeat Customers with P-not-C=1: {pnotc_1_percent:.2f}%")

plot_data = pd.DataFrame({
    'Category': ['cancelled', "didnt'cancel"],
    'Percentage': [pc_1_percent, pnotc_1_percent]
})

plt.figure(figsize=(7, 5))
sns.barplot(x='Category', y='Percentage', data=plot_data, palette=['black', 'blue'])

plt.title('Percentage of Repeat Customers (P-C=1 vs P-not-C=1)', fontsize=14)
plt.ylabel('Percentage (%)', fontsize=10)
plt.xlabel('') # Keep x-axis label simple
plt.ylim(0, max(plot_data['Percentage']) * 1.2 if not plot_data.empty else 100)
plt.show()

```

```python

print("Slide 4: Impact of Lead Time on Cancellations")

bins = [-1, 30, 90, 180, df['lead time'].max()] 
labels = ['0-30 Days', '31-90 Days', '91-180 Days', '180+ Days']
df['Lead_Time_Category'] = pd.cut(df['lead time'], bins=bins, labels=labels, right=True)

cancellation_by_lead_time = df.groupby('Lead_Time_Category')['booking status'].value_counts(normalize=True).unstack()

cancellation_rate_per_category = cancellation_by_lead_time.get('Canceled', pd.Series(0, index=labels)) * 100

print("\nCancellation Rate (%) by Lead Time Category:")
print(cancellation_rate_per_category.round(2))

plt.figure(figsize=(10, 6))
sns.barplot(x=cancellation_rate_per_category.index, y=cancellation_rate_per_category.values, palette='Blues_d')

plt.title('Cancellation Rate by Lead Time ', fontsize=16)
plt.xlabel('Lead Time Category', fontsize=12)
plt.ylabel('Cancellation Rate (%)', fontsize=12)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7) 
plt.show()

```

```python

print(" Booking Status by Market Segment Type")

segment_counts = df.groupby(['market segment type', 'booking status']).size().unstack(fill_value=0)

segment_counts['Canceled'] = segment_counts.get('Canceled', 0)
segment_counts['Not_Canceled'] = segment_counts.get('Not_Canceled', 0)

segment_counts['Cancellation Rate'] = (
    segment_counts['Canceled'] / (segment_counts['Canceled'] + segment_counts['Not_Canceled'] )
) * 100
segment_counts['Cancellation Rate'].fillna(0, inplace=True) # Handle potential NaN from division by zero

# Sort by cancellation rate 
segment_counts = segment_counts.sort_values(by='Cancellation Rate', ascending=False)

print("Booking Counts and Cancellation Rate by Market Segment Type:")
print(segment_counts[['Canceled', 'Not_Canceled', 'Cancellation Rate']].round(2))

# Create and display the bar chart
plt.figure(figsize=(9, 6))
sns.barplot(x=segment_counts.index, y=segment_counts['Cancellation Rate'], palette='viridis')

plt.title('Cancellation Rate by Market Segment Type')
plt.xlabel('Market Segment Type')
plt.ylabel('Cancellation Rate (%)')
plt.ylim(0, 100)
plt.xticks(rotation=45, ha='right')
plt.show()

```

```python

print("Cancellation Rate for Weekend vs. Week Nights")

def categorize_nights(row):
    has_weekend = row['number of weekend nights'] > 0
    has_week = row['number of week nights'] > 0

    if has_weekend and not has_week:
        return 'Weekend Only'
    elif not has_weekend and has_week:
        return 'Week Nights Only'
    else:
        return 'Both (Weekend & Week Nights)'

df['Night_Type_Category'] = df.apply(categorize_nights, axis=1)

cancellation_by_type = df.groupby('Night_Type_Category')['booking status'].value_counts(normalize=True).unstack(fill_value=0)

cancellation_by_type['Canceled'] = cancellation_by_type.get('Canceled', 0)
cancellation_by_type['Not_Canceled'] = cancellation_by_type.get('Not_Canceled', 0)

cancellation_by_type['Cancellation_Rate'] = (
    cancellation_by_type['Canceled'] / (cancellation_by_type['Canceled'] + cancellation_by_type['Not_Canceled'])
) * 100
cancellation_by_type['Cancellation_Rate'].fillna(0, inplace=True)

cancellation_by_type = cancellation_by_type.sort_values(by='Cancellation_Rate', ascending=False)

print("\nCancellation Rate by Night Type Category:")
print(cancellation_by_type[['Canceled', 'Not_Canceled', 'Cancellation_Rate']].round(2))

plt.figure(figsize=(9, 6)) 
sns.barplot(x=cancellation_by_type.index, y=cancellation_by_type['Cancellation_Rate'], palette='viridis')

plt.title('Cancellation Rate by Type of Nights Booked')
plt.xlabel('Booking Night Category')
plt.ylabel('Cancellation Rate (%)')
plt.ylim(0, 100)
plt.show()
```
