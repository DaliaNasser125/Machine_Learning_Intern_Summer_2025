```python
numeric_cols=['lead time', 'average price ',"P-not-C","P-C","number of week nights" , "number of weekend nights","repeated"]

def remove_outliers_iqr_all(data, columns):
    df_clean = data.copy()
    for col in columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df_clean = df_clean[(df_clean[col] >= lower) & (df_clean[col] <= upper)]
    return df_clean
df_cleaned = remove_outliers_iqr_all(data, numeric_cols)
```

```python
sns.boxplot(data= df_cleaned , palette="rainbow" , orient='h')
```
