```python
# @title Data Cleaning
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df
```

```python
df = df.rename(columns={'p-c': 'Preservation_Canceled'})
df = df.rename(columns={'p-not-c': 'Preservation_not_Canceled'})
df = df.rename(columns={'repeated': 'Visited_Before'})
df.columns
df
```

```python
# @title Univariate Analysis – Categorical Data
categorical_cols = ['type_of_meal', 'room_type', 'market_segment_type', 'booking_status']

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, col in enumerate(categorical_cols):
    sns.countplot(x=col, data=df, ax=axes[i], palette='pastel')
    axes[i].set_title(f'Distribution of {col}')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# @title Univariate Analysis – Numerical Data
numerical_cols = ['number_of_adults', 'number_of_children', 'lead_time', 'average_price', 'special_requests']

df[numerical_cols].hist(bins=20, figsize=(14, 10), color='skyblue')
plt.suptitle("Histograms of Numerical Features")
plt.show()
```

```python
# @title Bivariate Analysis
bivariate_cols = ['type_of_meal', 'room_type', 'market_segment_type']

# Create a 1x3 grid of subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for i, col in enumerate(bivariate_cols):
    sns.countplot(x=col, hue='booking_status', data=df, ax=axes[i], palette='Set2')
    axes[i].set_title(f'Booking Status by {col}')
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].set_xlabel(col.replace('_', ' ').title())

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='booking_status', y='average_price', data=df)
plt.title('Price Distribution by Booking Status')
plt.show()
```

```python
# @title Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# @title Feature Engineering
df['date_of_reservation'] = pd.to_datetime(df['date_of_reservation'], format='mixed', dayfirst=True, errors='coerce')
df['reservation_year'] = df['date_of_reservation'].dt.year
df['reservation_month'] = df['date_of_reservation'].dt.month
df['total_nights'] = df['number_of_weekend_nights'] + df['number_of_week_nights']
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='reservation_year', data=df, hue='reservation_year', palette='Blues', legend=False)
plt.title('Number of Reservations per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='reservation_month', data=df, hue='reservation_month', palette='Purples', legend=False)
plt.title('Number of Reservations per Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['total_nights'], bins=30, kde=True, color='orange')
plt.title('Distribution of Total Nights Stayed')
plt.xlabel('Total Nights')
plt.ylabel('Frequency')
plt.show()
```

```python
# @title Insights & Patterns
```

1. Most reservations are made during the summer months (september - october), peaking in October.

```python
monthly_counts = df['reservation_month'].value_counts().sort_index()
monthly_counts.plot(kind='bar', title='Reservations per Month', color='skyblue')
plt.xlabel('Month')
plt.ylabel('Reservations')
plt.show()

print("Monthly Reservation Counts:\n", monthly_counts)
```

2. 2018 had the highest number of reservations with 29,730 bookings, accounting for ~82% of the dataset.

```python
yearly_counts = df['reservation_year'].value_counts().sort_index()
yearly_counts.plot(kind='bar', title='Reservations per year', color='darkblue')
plt.xlabel('year')
plt.ylabel('Reservations')
plt.show()

print("Yearly Reservation Counts:\n", yearly_counts)
```

3. Weekend stays are generally shorter, with most guests staying 0 to 2 weekend nights. Over 98% of the bookings include 2 or fewer weekend nights.

```python
weekend_counts = df['number_of_weekend_nights'].value_counts().sort_index()

plt.figure(figsize=(8, 5))
sns.barplot(x=weekend_counts.index, y=weekend_counts.values, color='skyblue')
plt.title('Distribution of Weekend Night Stays')
plt.xlabel('Number of Weekend Nights')
plt.ylabel('Number of Bookings')

short_stays = df[df['number_of_weekend_nights'] <= 2].shape[0]
total_bookings = df.shape[0]
percentage_short_stays = (short_stays / total_bookings) * 100

plt.text(2.5, max(weekend_counts.values) * 0.9,
         f"{percentage_short_stays:.2f}% of bookings\nhave 0–2 weekend nights",
         fontsize=12, color='darkblue', bbox=dict(facecolor='lightyellow', edgecolor='blue'))

plt.tight_layout()
plt.show()
```

4. Room_Type 1 is by far the most frequently booked room, accounting for 77.5% of all bookings (28,138 out of 36,285).

```python
# Count of room types
most_common_room = df['room_type'].value_counts()
cancel_by_room = df.groupby('room_type')['Preservation_Canceled'].mean().sort_values()

plt.figure(figsize=(6, 4))
sns.barplot(x=most_common_room.index, y=most_common_room.values)
plt.title('Most Common Room Types')
plt.xlabel('Room Type')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

most_common_room = df['room_type'].value_counts()
most_common_room_percent = (most_common_room / df.shape[0]) * 100

print("Most Common Room Types (Counts & Percentages):\n")
for room, count in most_common_room.items():
    percent = most_common_room_percent[room]
    print(f"{room}: {count} bookings ({percent:.2f}%)")
```

5. Bookings from the "Online" market segment have the highest cancellation rate (~33%)

```python
cancel_rate = df.groupby('market_segment_type')['booking_status'].apply(lambda x: (x == 'Canceled').mean()).sort_values(ascending=False)
cancel_rate.plot(kind='bar', title='Cancellation Rate by Market Segment')
plt.show()

cancel_rate = (df['booking_status'] == 'Canceled').mean()
print(f'Overall Cancellation Rate = {cancel_rate*100:.2f}%')
```

6. Room_Type 7 has the highest cancellation rate at ~21%

```python
plt.figure(figsize=(6, 4))
sns.barplot(x=cancel_by_room.index, y=cancel_by_room.values)
plt.title('Cancellation Rate by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Cancellation Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

cancel_by_room = df.groupby('room_type')['Preservation_Canceled'].mean().sort_values()
cancel_by_room_percent = cancel_by_room * 100

print("\nCancellation Rate by Room Type (%):\n")
for room, cancel_rate in cancel_by_room_percent.items():
    print(f"{room}: {cancel_rate:.2f}% cancellation rate")
```

7. Guests who make special requests are significantly less likely to cancel.

```python
sns.barplot(x='special_requests', y='Preservation_Canceled', data=df)
plt.title('Cancellation Rate by Number of Special Requests')
plt.xlabel('Special Requests')
plt.ylabel('Cancellation Rate')
plt.show()
```
